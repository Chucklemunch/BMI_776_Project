```python
import pandas as pd
import numpy as np
import pickle
import scanpy as sc
import sys
from matplotlib import pyplot as plt
```

```python
with open('../Data/SARS-CoV-2/gene_expression_matrix.tsv', 'r') as f:
    data = pd.read_table(f)

data
```

```python
data = data.set_index('gene_id')
data
```

```python
obs = data.columns.values
obs
```

```python
# Get raw expression count matrix and save to file for use in GRN algos
exp_matrix = data.to_numpy().T

with open('../Data/SARS-CoV-2/tpm_exp_matrix.pickle', 'wb') as f:
    pickle.dump(exp_matrix, f)
```

```python
# Get highly variable genes 

# Creating AnnData Object from
adata = sc.AnnData(X=exp_matrix, obs=obs)

# log-transforming data so we can find highly variable genes
sc.pp.log1p(adata)

# Find highly variable genes
sc.pp.highly_variable_genes(adata, n_top_genes=500)

# Getting top 500 most highly variable genes
highly_variable_idx = [i for i in range(adata.X.shape[1]) if adata.var['highly_variable'].iloc[i] == True]

highly_variable_genes = data.index[highly_variable_idx].to_list()

# Save list of top 500 (503) highly variable genes
with open('../Data/SARS-CoV-2/highly_variable_genes_503.pickle', 'wb') as f:
    pickle.dump(highly_variable_genes, f)

# Subsetting expression data to highly variable genes
highly_variable_genes_matrix = data.loc[highly_variable_genes].to_numpy()

# Outputting matrix of highly variable genes for use in GRN algos
with open('../Data/SARS-CoV-2/highly_variable_expression_matrix_503.pickle', 'wb') as f:
    pickle.dump(highly_variable_genes_matrix, f)

highly_variable_genes
```
